```python
# Author: Nicholas Makarov
# Date 3/23/2022
# Comments: This program uses python to replicate PROGRAM (4) from Deliverable (5)



import numpy as np

def populate_list(length):
    # Generates a list with random integers between 10 and 50
    return np.random.randint(10, 51, length)

def sum_list(lst):
    # Sum all the elements in a list
    return sum(lst)

# Ask the user to enter an integer number between 5 and 15.
while True:
    try:
        user_input = int(input("Enter an integer number between 5 and 15: "))
        if 5 <= user_input <= 15:
            break
        else:
            print("Invalid input. Please enter an integer number between 5 and 15.")
    except ValueError:
        print("Invalid input. Please enter an integer number between 5 and 15.")

# Use Method 1 
lst = populate_list(user_input)

# Display in the console all the elements the in the list.
print("The elements of the arrays are:", *lst)

# Use Method 2 
total = sum_list(lst)

# Display in the console the value from step 5.
print("The sum is:", total)
```
